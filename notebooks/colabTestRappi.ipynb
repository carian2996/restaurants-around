```
import requests
import json
```

```
check_coverage = 'https://services.mxgrability.rappi.com/api/base-crack/has-coverage?lat={}&lng={}'
```

```
ADDRESS = 'Avenida Revolución 258, Tacubaya, Mexico City, CDMX'
LAT = 19.3987204
LNG = -99.184695
```

```
get_coverage = requests.get(check_coverage.format(LAT, LNG))
```

```
if json.loads(get_coverage.text)['has_coverage']:
    print('Has coverage')
```

```
catalog = 'https://services.mxgrability.rappi.com/api/restaurant-bus/stores/catalog/home'
```

```
post_catalog = requests.post(catalog, json={"lat": LAT, "lng": LNG, "store_type": "restaurant", "is_prime": "false"})
```

```
dict_catalog = json.loads(post_catalog.text)
```
