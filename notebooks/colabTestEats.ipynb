```
# I/O modules
import csv
import json

# Scraping tools
from selenium.webdriver.support.ui import Select, WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium import webdriver

# Utilities
import time
import re
```

```
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
```

```
driver = webdriver.Chrome('chromedriver', options=chrome_options)
```

```
URL = 'https://www.ubereats.com/mx'
# ADDRESS = 'Avenida Revolución 258, Tacubaya, Mexico City, CDMX'
ADDRESS = 'Nicolas Leon 173, Jardin Balbuena, Mexico City, CDMX'
```

```
elems = driver.find_elements_by_xpath("//a[@href]")
driver_cat_hrefs = {}
for elem in elems:
    url = elem.get_attribute("href")
    if 'near-me/' in url: 
        href_txt = url.replace('https://www.ubereats.com', '')
        cat = href_txt.replace('/mx/near-me/', '')
        # print('Looking restaurants for:', cat)
        print(href_txt)
        driver_cat_hrefs[cat] = href_txt # driver.find_element_by_xpath("//a[@href='" + href_txt + "']")
```
